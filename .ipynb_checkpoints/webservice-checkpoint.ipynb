```python
import numpy as np

#parser
import pandas as pd

#plot
import matplotlib.pyplot as plot

#tensorflow
import tensorflow.keras as kr
```

```python
data = pd.read_csv("powerproduction.txt")

plot.scatter(
    data.values[:,0], data.values[:,1],
    s=10, c='black')


#print dataset
plot.xlabel("speed")
plot.ylabel("power")
plot.grid()
plot.show()
```

```python
#neural network
model = kr.models.Sequential()
model.add(kr.layers.Dense(50, input_shape=(1,), activation="sigmoid", kernel_initializer='ones', bias_initializer='zeros'))
model.add(kr.layers.Dense(1, input_shape=(1,), activation="linear", kernel_initializer='ones', bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')

#train
model.fit(data['speed'], data['power'], epochs=1000);
```

```python
plot.plot(data['speed'], data['power'], 'r.', label='data')
plot.plot(data['speed'], model.predict(data['speed']), color='blue', label='predicted')
plot.legend();
```

```python
#loss
model.evaluate(data['speed'], data['power'])
```
