```python
#imports
import numpy as np
#parser
import pandas as pd
#plot
import matplotlib.pyplot as plt
#tensorflow
import tensorflow.keras as kr
#dbscan for outlier detection
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler


```

```python
data = pd.read_csv("powerproduction.txt")

plt.scatter(data.values[:,0] , data.values[:,1] ,
    s=10, c='black')

#print dataset
plt.xlabel("speed")
plt.ylabel("power")
plt.grid()
plt.show()
```

```python

#neural network
model = kr.models.Sequential()
model.add(kr.layers.Dense(25, input_shape=(1,), activation="sigmoid", kernel_initializer='ones', bias_initializer='zeros'))
model.add(kr.layers.Dense(25, input_shape=(1,), activation="sigmoid", kernel_initializer='ones', bias_initializer='zeros'))
model.add(kr.layers.Dense(1, input_shape=(1,), activation="linear", kernel_initializer='ones', bias_initializer='zeros'))
model.compile('adam', loss='mean_squared_error')

#train
model.fit(data.values[:,0] , data.values[:,1] , epochs=500,batch_size=10)
```

```python
#loss
model.evaluate(data.values[:,0] , data.values[:,1])
```

Here we have evaluated our neural network, our loss is approx 249,
this is not ideal as we want to try and get a loss down as far as low decimals.
Outlier pieces of data can be seen on the X axis below which I believe to be the cause of 
inaccuracey.<br/>
The red line Shows the predicted curve for power Output


```python
#show the predict 
plt.plot(data.values[:,0] , data.values[:,1] ,('.g') , label='data')
plt.plot(data.values[:,0], model.predict(data.values[:,0]), color='red',label='Predicted Output')
plt.legend();
```
